# Chemotherapy

```python
from tumortwin.types.treatment import ChemotherapySpecification
from models.test_chemotherapy import test_chemotherapy_model


U_0 = 0.5
chemo_specs = [
    ChemotherapySpecification(sensitivity=0.01, decay_rate=0, protocol={0.0: 1.0})
]
test_chemotherapy_model(U_0, chemo_specs, plot=True)
chemo_specs = [
    ChemotherapySpecification(sensitivity=0.2, decay_rate=0.9, protocol={0.0: 1.0})
]
test_chemotherapy_model(U_0, chemo_specs, plot=True)
chemo_specs = [
    ChemotherapySpecification(sensitivity=0.2, decay_rate=0.9, protocol={2.5: 1.0})
]
test_chemotherapy_model(U_0, chemo_specs, plot=True)
chemo_specs = [
    ChemotherapySpecification(sensitivity=0.2, decay_rate=0.9, protocol={0.0: 0.1}),
    ChemotherapySpecification(sensitivity=0.2, decay_rate=0.9, protocol={2.5: 1.0}),
]
test_chemotherapy_model(0.5 * U_0, chemo_specs, plot=True)
```
